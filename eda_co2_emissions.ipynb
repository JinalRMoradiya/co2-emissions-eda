```python
# Remove quotes from column names
df.columns = [col.strip('"') for col in df.columns]

# Verify the change
df.columns
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month", y="1958", data=df, marker="o")
plt.title("Monthly CO2 Data - 1958")
plt.xlabel("Month")
plt.ylabel("Value")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month", y="1958", data=df, marker="o")
plt.title("Monthly CO2 Data - 1958")
plt.xlabel("Month")
plt.ylabel("Value")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(x="Month", y="1958", data=df, marker="o", label="1958")
sns.lineplot(x="Month", y="1959", data=df, marker="o", label="1959")
sns.lineplot(x="Month", y="1960", data=df, marker="o", label="1960")

plt.title("Monthly CO2 Data Comparison")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.show()
```

# CO2 Emissions Exploratory Data Analysis

**Dataset:** Monthly CO2 data for 19581960  
**Goal:** Explore and visualize trends over time

## Summary of Findings
- CO2 levels generally increase during summer months (JuneAugust)
- 1960 shows higher overall values compared to 1958 and 1959
- Monthly patterns are consistent across years with seasonal peaks
